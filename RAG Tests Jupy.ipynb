```python
import os
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Configuración del modelo
model_name = "EleutherAI/gpt-neo-2.7B"
device = "mps" if torch.backends.mps.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# Esquema de la base de datos
esquema = """
Esquema de la base de datos:
Tabla: Transaccion
- instalacion: CharField, max_length=100
- fecha_hora: DateTimeField, auto_now_add=True
- ubicacion: CharField, max_length=100
- transaccion: CharField, max_length=3 (choices: 'INGRESO', 'RETIRO')
- codigo_articulo: CharField, max_length=50
- descripcion_articulo: TextField
- cantidad: IntegerField
- usuario: CharField, max_length=50
"""

def generar_query(prompt):
    # Incluir el esquema en el prompt
    full_prompt = f"{esquema}\n\nTexto natural: {prompt}\nConsulta SQL:"
    inputs = tokenizer(full_prompt, return_tensors="pt").to(device)

    # Generar la consulta SQL
    outputs = model.generate(inputs['input_ids'], max_length=150, num_return_sequences=1)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text.split("Consulta SQL:")[-1].strip()  # Extraer solo la consulta SQL

# Ejemplo de uso
prompt = "Genera un query SQL para obtener todas las transacciones de la instalación 'A'."
query = generar_query(prompt)
print("Consulta SQL generada:", query)
```

```python
def generar_query(prompt):
    full_prompt = f"{esquema}\n\nTexto natural: {prompt}\nConsulta SQL:"
    inputs = tokenizer(full_prompt, return_tensors="pt").to(device)

    # Set the pad token ID if not already set
    if tokenizer.pad_token_id is None:
        tokenizer.pad_token_id = tokenizer.eos_token_id

    # Generate the query with attention mask and pad token ID
    outputs = model.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        pad_token_id=tokenizer.pad_token_id,
        max_new_tokens=103,
        num_return_sequences=1
    )
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text.split("Consulta SQL:")[-1].strip()
```
